```python
def transform_pd_to_ad(df, feature_cols, meta_obs):
    features = df[[c for c in feature_cols]].to_numpy()
    adata = ad.AnnData(features)
    adata.obs_names = [f"Cell_{i:d}" for i in range(adata.n_obs)]
    adata.var_names = feature_cols
    for k,v in meta_obs.items():
        adata.obs[k] = pd.Categorical(v)
```
